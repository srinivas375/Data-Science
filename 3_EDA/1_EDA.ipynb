**Changing the dataypes of *columns***

```python
amazon['discounted_price'] = amazon['discounted_price'].str.replace('₹','')
amazon['discounted_price'] = amazon['discounted_price'].str.replace(',','')
amazon['discounted_price'] = amazon['discounted_price'].astype(float)

amazon['actual_price'] = amazon['actual_price'].str.replace('₹','')
amazon['actual_price'] = amazon['actual_price'].str.replace(',','')
amazon['actual_price'] = amazon['actual_price'].astype(float)
```

```python
# changing the dtype and values of percentage

amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%','').astype('float64')
amazon['discount_percentage'] = amazon['discount_percentage']/100
```

```python
# value of counts of ratings

amazon['rating'].value_counts()
```

```python
# checking that row

amazon[amazon['rating'] == "|"]
```

```python
amazon.query("rating == '|'")
```

row 1279 is erronous

```python
# replacing the erronous rating with mean rating value

mean_rating = amazon['rating']
mean_rating = mean_rating.str.replace('|','0')
mean_rating = mean_rating.astype(float)
mean_rating = mean_rating.mean()
print(mean_rating)
```

```python
amazon['rating'] = amazon['rating'].replace('|',str(mean_rating)).astype('float64')
```

```python
# changing the "rating_count" column type

amazon['rating_count'] = amazon['rating_count'].str.replace(',','').astype('float64')
```

```python
amazon.info()
```

```python
# finding the descriptive statistics

amazon.describe()
```

```python
amazon.isnull().sum().sort_values(ascending=False)
```

```python
# viewing the null valued data

amazon[amazon.isnull().any(axis=1)]
```

```python
# imputing the missing values of rating_count with median value

amazon['rating_count'] = amazon['rating_count'].fillna(amazon['rating_count'].median())
```

## **Analysis of the data**

```python
# quality analysis of the wine data

quality_counts = wine['quality'].value_counts()
fig = px.bar(quality_counts, x=quality_counts.index, y=quality_counts.values,color = quality_counts.index, title = "Quality analysis of the wine data")
fig.show()
```

This plot shows the quality rate of the wine data

```python
# kernal plot to understand the variance in the data

sns.set_theme(style = "darkgrid")

cols = wine.select_dtypes(include=['int64','float64']).columns

plt.figure(figsize=(14, len(cols)*3))

for i, feature in enumerate(cols, 1):
  plt.subplot(len(cols), 2, i)
  sns.histplot(data=wine, x=feature, kde=True)
  plt.title(f"{feature} | skewness: {round(wine[feature].skew(), 2)}")

plt.tight_layout()
plt.show()


```

```python
# sworm plot to identify the outliers

plt.figure(figsize=(10,8))
sns.swarmplot(data = wine, x = "quality", y = "alcohol", hue = "quality")
plt.title("Alcohol vs Quality")
plt.show()
```

```python
# violin plot for same analysis

fig = px.violin(wine, x='quality', y='alcohol', title = "Alcohol vs Quality", color = "quality")
fig
```

```python
# box plot for outlier analysis

fig = px.box(wine, x='quality', y='alcohol',color = 'quality', title = "Alcohol vs Quality")
fig
```

```python
# finding the correlation b/w the data

plt.figure(figsize=(15,10))
sns.heatmap(wine.corr(), annot=True, linewidths=2, fmt='.2f', cmap = 'Pastel2')
plt.title("correlation heatmap")
plt.show()
```

Loading and observing the data of titanic dataset

```python
# observing the columns

titanic.columns.tolist()
```

```python
# unique values of each column

titanic.nunique()
```

What does each column says
* **PassengerId**: id of the passenger
* **Survived**: Wether passenger is survived or not (1/0)
* **Pclass**: Travelling class of passenger (1/2/3)
* **Name**: name of the passenger
* **Sex**: gender of the passenger
* **Age**: age of the passenger
* **Sisb**: how many related members travelling with that passenger (count)
* **Parch**: how many parents or childs travelling with that passenger (count)
* **Ticket**: ticket number
* **Fare**: what is the cost of ticket
* **cabin**: in which cabin does the customer travells
* **Embarked**: From which station does this titanic starts travelling (S/C/QT)
