### **mode, count, proportion**

```python
for col in categorical_cols:
    values = data[col]
    mode_val = st.mode(values)
    print(f"Statistics for {col}:")
    print(f"Mode: {mode_val}")
    value_counts = Counter(values)
    print(f"Value counts: {value_counts}")
    total_responses = len(values)
    proportions = {value : count/total_responses for value, count in value_counts.items()}
    print(f"Proportion: {proportions}\n")
```

### **Summary measures for numerical variables**

```python
# defining the numerical columns

numerical_cols = ['Age', 'Time_Spent', 'No_of_Devices', 'Monthly_OTT_Bill']
```

#### **Measures of central tendency**

**Mean, median, mode** are the central tendecy for numerical variables

```python
for col in numerical_cols:
    values = data[col]
    mode_val = st.mode(values)
    print(f"Statistics for {col}:")
    print(f"Mode: {mode_val}")
    mean_val = st.mean(values)
    print(f"Mean: {mean_val}")
    median_val = st.median(values)
    print(f"Median: {median_val}\n")
```

#### **Measures of variation**

Calculating the **range, IQR, var, std, cv(Coefficient of Variation)**

```python
for col in numerical_cols:
    values = data[col]
    range_val = np.ptp(values)   # max - min
    q3, q1 = np.percentile(values, [75, 25])
    iqr = q3 - q1
    stdiv = np.std(values)
    variance = np.var(values)
    mean = np.mean(values)
    cv = (stdiv/mean)
    print(f"Statistics of {col}:")
    print(f"Range: {range_val:.2f}")
    print(f"IQR: {iqr:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"standard diviation: {round(stdiv,2)}")
    print(f"Coefficient of Variation: {cv:.2f}\n")
```

```python
# mean of sampling means is equal to population mean

population_mean = data['Age'].mean()

n_iterations = 1000
sample_size = 10

sample_means = []
for _ in range(n_iterations):
    sample = data['Age'].sample(n=sample_size, replace=False)
    sample_means.append(sample.mean())

sample_means = np.array(sample_means)

print(f"Population Mean: {population_mean:.4f}")
print(f"Mean of Sample Means: {sample_means.mean():.4f}")
print(f"Difference: {abs(population_mean - sample_means.mean()):.6f}\n")
print(f"first 20 Sample Means: {sample_means[:20]}")
```

### **plotting histograms of sample mean**

```python
from scipy.stats import norm
```

```python
min_sample_mean = np.min(sample_means)
max_sample_mean = np.max(sample_means)

mean_sample_mean = np.mean(sample_means)
std_sample_mean = np.std(sample_means)
x = np.linspace(min_sample_mean, max_sample_mean, 100)
y = norm.pdf(x, mean_sample_mean, std_sample_mean)

plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.4, density=True)
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
plt.axvline(sample_means.mean(), color='yellow', linestyle='dotted', linewidth=2, label='Mean of Sample Means')
plt.plot(x,y, label = "Normal Distribution", ls = '--', color = 'orange')
plt.title('Sampling Distribution of the Sample Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

We can observe that, sample means distribution is nearly to normal

## **Central Limit Theorem**

For large enough samples, the sampling distribution of the mean is approximately normal, even if the population distribution isn’t.

```python
# lets see the population distribution of age again

sns.histplot(data['Age'], kde=True, bins = 30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

We can observe that, the distribution of age is **not Normal**

```python
sample_sizes = [5, 10, 30, 50, 70, 100]
n_iterations = 2000
```

```python
plt.figure(figsize=(15, 10))

for i, size in enumerate(sample_sizes, 1):
    sample_means = []
    
    for _ in range(n_iterations):
        sample = data['Age'].sample(n=size, replace=False)
        sample_means.append(sample.mean())
    
    sample_means = np.array(sample_means)
    
    print(f"Sample size = {size}")
    print(f"Mean of Sample Means: {sample_means.mean():.4f}")
    print(f"Population Mean: {population_mean:.4f}")
    print(f"Difference: {abs(population_mean - sample_means.mean()):.6f}")
    print(f"Std Dev of Sample Means: {sample_means.std(ddof=1):.4f}\n")
    
    plt.subplot(3, 3, i)
    plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7)
    plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
    plt.title(f'Sample size = {size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.legend()

plt.tight_layout()
plt.show()
```

We can observe that,
* As the sample size incereases, the "mean of sample mean is more nearly to population mean"
* As the sample size incerases, the wideness also decreases (values of standard deviations proves that statement)

```python
# proving the statement 2

for i, size in enumerate(sample_sizes, 1):
    sample_means = []
    
    for _ in range(n_iterations):
        sample = data['Age'].sample(n=size, replace=False)
        sample_means.append(sample.mean())
    
    sample_means = np.array(sample_means)
    
    plt.hist(sample_means, bins = 30, alpha = 0.6, label = f"sample size: {size}")

plt.xlabel("sample mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem visualization")
plt.legend()
plt.show()
```

As the sampling size incerases, the curve becomes more normal and the width also decreases

### **sample distribution of sample variance**

```python
from scipy.stats import chi2, norm
```

```python
sample_size = 140
num_samples = 1000
```

```python
population_variance = np.var(data['Age'])

sample_variances = []
for i in range(num_samples):
    sample = data.sample(sample_size, replace=False)
    sample_var = sample['Age'].var()
    sample_variances.append(sample_var)

min_sample_variance = min(sample_variances)
max_sample_variance = max(sample_variances)

mean_sample_variance = np.mean(sample_variances)
std_sample_variance = np.std(sample_variances)

x_norm = np.linspace(min_sample_variance, max_sample_variance, 1000)
y = norm.pdf(x_norm, mean_sample_variance, std_sample_variance)

# chi square 
dof = sample_size - 1
x_vals = np.linspace(min(sample_variances), max(sample_variances), 1000)
y_chi = chi2.pdf((dof) * x_vals / population_variance, df=dof) * (dof / population_variance)

plt.hist(sample_variances, bins = 30, range=(min_sample_variance, max_sample_variance), density=True, alpha = 0.6, label="Sample var's dist", edgecolor = 'k')
plt.plot(x_norm, y, 'r--', label = 'Normal distribution')
plt.plot(x_vals, y_chi, 'g:', label = 'Chi square distribution')
plt.title("Sample variances distribution")
plt.grid()
plt.legend()
plt.xlabel('sample variance')
plt.ylabel("Density")
plt.show()
```

### **Binomial distribution**

```python
from scipy.stats import binom
```

```python
params = [
    (20, 0.1),
    (20, 0.2),
    (20, 0.3),
    (20, 0.4),
    (20, 0.5),
    (20, 0.6),
    (20, 0.7),
    (20, 0.8),
    (20, 0.9),
]

plt.figure(figsize=(15, 8))

for i, (n, p) in enumerate(params, 1):
    x = np.arange(0, n+1)
    
    pmf_vals = binom.pmf(x, n, p)
    
    plt.subplot(2, 5, i)
    plt.vlines(x, 0, pmf_vals, colors='b', lw=2)
    plt.plot(x, pmf_vals, 'bo', ms=8)  # dots on top of lines
    
    plt.title(f"Binomial PMF\nn={n}, p={p}")
    plt.xlabel("Number of successes")
    plt.ylabel("Probability")
    plt.ylim(0, max(pmf_vals) + 0.05)
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

As the p value changes, the distribution of the curve also changes

```python
params = [
    (10, 0.5),
    (30, 0.5),
    (50, 0.5)
]

colors = ['g', 'r', 'b']

for i, (n, p) in enumerate(params, 1):
    x = np.arange(0, n+1)
    
    pmf_vals = binom.pmf(x, n, p)
    
    plt.vlines(x, 0, pmf_vals, colors=colors[i-1], lw=2, label=f"n: {n}, p: {p}")
    plt.plot(x, pmf_vals, 'bo', ms=8, color = colors[i-1])
    plt.grid(alpha=0.3)


plt.title("Binomial distribution")
plt.xlabel("Number of successes")
plt.ylabel("Probabibility")
plt.legend()
plt.show()
```
