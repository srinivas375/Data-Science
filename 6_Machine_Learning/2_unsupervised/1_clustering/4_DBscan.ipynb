## **DBscan**

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
X, y = make_blobs(
    n_samples=500,
    centers = 4,
    cluster_std = 1.2,
    random_state = 42
)
```

```python
# plotting

sns.scatterplot(x=X[:, 0], y=X[:, 1])
plt.title("simple example")
plt.show()
```

```python
dbscan = DBSCAN(
    eps=1,
    min_samples = 3
)
labels = dbscan.fit_predict(X)
```

```python
if len(set(labels))>1 and -1 not in set(labels):
  score = silhouette_score(X, labels)
  print(f"Silhouette Score: {score:.3f}")
else:
  if -1 in set(labels):
    print("Silhouette score not defined (noise present)")
  else:
    print("Silhouette score not defined (only 1 cluster)")
```

```python
# plotting

unique_labels = set(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))

plt.figure(figsize=(8, 6))
for k, col in zip(unique_labels, colors):
  if k == -1:
    col = 'k'
    label_name = 'Noise'
  else:
    label_name = f'Cluster {k}'

  class_member_mask = (labels == k)
  xy = X[class_member_mask]
  plt.scatter(xy[:, 0], xy[:, 1], c=[col], label=label_name, s=30, edgecolors='k')

plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

* Including the predicting the data clusters, it also predicts the noise points
* DBscan can handle most of the different distributions of data
