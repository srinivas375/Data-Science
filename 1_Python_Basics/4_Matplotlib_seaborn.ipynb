# **Matplotlib**

**Matplotlib** is a popular Python library used for data visualization. It allows you to create a wide variety of **static, animated, and interactive plots**, such as line **graphs, bar charts, histograms, and scatter plots**.  
The most commonly used module in Matplotlib is **pyplot**, which provides a simple interface for plotting.

Installing libraries:
* Using pip:
    - `pip install matplotlib seaborn`
* Using uv:
    - `uv add matplotlib seaborn`

## **Importing and labeling**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
```

```python
# Creating a line plot
plt.plot(x, y)

# Adding labels and title
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


# saving the generated plot
plt.savefig("basic_line_plot.png")

# Show the plot
plt.show()
```

## Plot types

### **`Line plot`**

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### **`Bar chart`**



```python
categories = ['A', 'B', 'C']
values = [10, 25, 18]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

### **`Horizontal Bar Chart`**

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()
```

### **`Scatter Plot`**



```python
x = np.linspace(0, 10, 10)
y = x ** 2

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### **`Pie Chart`**

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

### **`Histogram`**

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## **Customizations**

### **For line plot**

Change Line Color, Style, and Marker

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.plot(x, y, color='green', linestyle='--', marker='o', mec = 'black', mfc = 'yellow')
plt.title("Trading chart", loc = 'center')        # location can be 'center', 'left', 'right'
plt.xlabel("Day")
plt.ylabel("Trade value")
plt.show()
```

Add Legend

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 10, 20]

plt.plot(x, y1, label="amazon", color='blue')
plt.plot(x, y2, label="google", color='orange')
plt.title("Trading chart")
plt.xlabel("Day")
plt.ylabel("Trade value")
plt.legend()
plt.show()
```

Add Grid Lines

```python
plt.plot(x, y1)
plt.title("Grid Example")
plt.grid(True)
plt.show()
```

Set Axis Limits -- like zooming graph

```python
plt.plot(x, y1)
plt.title("Axis Limits")
plt.xlim(0, 5)
plt.ylim(0, 35)
plt.show()
```

### **For Scatter plot**

color to points

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, c="red")
plt.title("Scatter Plot with Color")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

with color map

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x, y, c=colors, cmap='viridis')
plt.title("Scatter Plot with Color Map")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## **Multiple plots in one plot**

Multiple lines in one plot

```python
x = [1, 2, 3, 4, 5]
y1 = [12, 20, 35, 40, 55]
y2 = [5, 14, 20, 35, 40]
y3 = [49, 32, 23, 44, 5]

plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='red')
plt.plot(x, y3, label='Line 3', color='green')
plt.title("Multiple Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### **Subplots using **`plt.subplot()`****

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [20, 17, 23, 9]

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
# with mulitple types

x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [20, 17, 23, 9]

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("Plot 1 (line plot)")

plt.subplot(2, 1, 2)
plt.scatter(x, y2)
plt.title("Plot 2 (scatter plot)")

plt.tight_layout()
plt.show()
```

```python
# with mulitple types

x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [20, 17, 23, 9]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1 (line plot)")

plt.subplot(2, 2, 2)
plt.scatter(x, y2)
plt.title("Plot 2 (scatter plot)")

plt.subplot(2, 2, 3)
plt.bar(x, y1)
plt.title("Plot 3 (bar chart)")

plt.subplot(2, 2, 4)
plt.barh(x, y2)
plt.title("Plot 4 (horizontal bar chart)")

plt.tight_layout()
plt.show()
```

### **Subplots using **`plt.subplot() with axes`****

```python
fig, axs = plt.subplots(2, 1, figsize = (8, 6))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

axs[0].plot(x, y, color='purple')
axs[0].set_title("Line Plot")

axs[1].bar(x, y, color='green')
axs[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

## **3D plotting**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

### **`3D Line Plot`**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(4)
y = np.arange(0, 7, 2)
z = [0, 3, 1, 5]

ax.plot(x, y, z, marker='o', linestyle='-')
ax.set_title("3D Line Plot")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

### **`3D Scatter plot`**


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_title("3D Scatter Plot")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

### **`3d Surface plot`**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

## **3D contour and 2D contour plots**

```python
#3d contour plot

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='viridis')
ax.set_title("3D Contour Plot")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
# corresponding 2d plot

plt.contour(X, Y, Z, 20, cmap='viridis')
plt.colorbar(label = "Z-value")
plt.title("2D Contour Plot")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## **Heatmap using matplotlib**

```python
data = np.random.rand(4, 4)*10

plt.imshow(data, cmap='hot', interpolation = 'nearest')
plt.colorbar(label='Value')
plt.title("Heatmap example")

rows, cols = data.shape
for i in range(rows):
    for j in range(cols):
        plt.text(j, i, f"{data[i, j]:.2f}", ha='center', va='center', color='green')
plt.show()
```

# **Searborn**

**Seaborn** is a python data visualization library built on top of **matplotlib**.  
It provides a high level interface for drawing **attractive, statistical graphics** with less code.

## **Seaborn datasets and configuration**

```python
import seaborn as sns
import pandas as pd
```

```python
# seeing the available datasets in seaborn

names = sns.get_dataset_names()
print(names)
```

```python
# loading a dataset, loaded in form of pandas dataframe

df = sns.load_dataset('tips')
print(type(df))
```

```python
df.head()
```

```python
# setting a theme for seaborn plots
# default theme is "white"

sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
```

```python
# using "whitegrid" theme

sns.set_theme(style = "whitegrid")
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
```

```python
# using "dark" theme

sns.set_theme(style = "dark")
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
```

```python
#using "white" theme

sns.set_theme(style = "white")
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
```

```python
# using "ticks" theme

sns.set_theme(style = "ticks")
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
```

```python
# using "darkgrid" theme

sns.set_theme(style = "darkgrid")
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title("Average total bill per day")
plt.show()
#
```

## **Categorical plots in seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### **`barplot()`**

```python
sns.barplot(data = tips, x = "day", y = "total_bill")
plt.title("Average total bill per day")
plt.show()
```

```python
sns.barplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.title("Average total bill per day, segregated by gender")
plt.show()
```

### **`countplot()`**

```python
sns.countplot(data = tips, x = "sex")
plt.title("Count of genders")
plt.show()
```

```python
sns.countplot(data = tips, x = "sex", hue = "day")
plt.title("Count of genders, segregated by day")
plt.show()
```

### **`boxplot()`**

```python
sns.boxplot(data = tips, x = "day", y = "total_bill")
plt.title("Boxplot of total bill per day")
plt.show()
```

```python
sns.boxplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.title("Boxplot of total bill per day, segregated by gender")
plt.show()
```

### **`violinplot()`**

```python
sns.violinplot(data = tips, x = "day", y = "total_bill")
plt.title("Violin plot of total bill per day")
plt.show()
```

```python
sns.violinplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.title("Violin plot of total bill per day, segregating by gender")
plt.show()
```

```python
sns.violinplot(data = tips, x = "day", y = "total_bill", hue = "sex", split = True)
plt.title("Violin plot of total bill per day, segregating by gender")
plt.show()
```

### **`spripplot()`**

```python
sns.stripplot(data = tips, x = "day", y = "total_bill")
plt.title("Stripplot of total bill per day")
plt.show()
```

### **`swarmplot()`** -- it is a stripplot but avoids overlap

```python
sns.swarmplot(data = tips, x = "day", y = "total_bill", hue = 'sex')
plt.title("Swarmplot of total bill per day")
plt.show()
```

```python
# scatter plot

sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", style = "sex", size = 'size')
plt.title("Scatterplot of total bill vs tip", loc = "center")
plt.show()
```
